```r
# Set the seed for reproducibility
set.seed(123)

# Load the data
X_train <- read.csv("/content/drive/My Drive/TDS data/X_trainTDS.csv")
X_test <- read.csv("/content/drive/My Drive/TDS data/X_testTDS.csv")
Y_train <- read.csv("/content/drive/My Drive/TDS data/Y_trainTDS.csv")
Y_test <- read.csv("/content/drive/My Drive/TDS data/Y_testTDS.csv")
```

```r
# Convert Y_train and Y_test to a vector
Y_train <- Y_train[, 1]
Y_test <- Y_test[, 1]

# Print the dimensions of X_train and X_test
print(dim(X_train))
print(dim(X_test))
```

```r
# Define function to plot confusion matrix
confusionMatrixPlot <- function(cm, title) {
  cm_table <- as.table(cm$table)
  cm_df <- as.data.frame(cm_table)
  colnames(cm_df) <- c("Reference", "Prediction", "Freq")

  ggplot(cm_df, aes(x = Prediction, y = Reference, fill = Freq)) +
    geom_tile(color = "white") +
    geom_text(aes(label = Freq), color = "black", size = 5) +
    scale_fill_gradient(low = "white", high = "blue") +
    ggtitle(title) +
    xlab("Predicted") +
    ylab("Actual") +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5))
}
```

```r
# Combine the training data
train_data <- data.frame(X_train, Class = Y_train)
```

```r
# Train the logistic regression model
model <- glm(Class ~ ., data = train_data, family = binomial)

# Predict class labels on the test set
y_pred_lr <- predict(model, newdata = X_test, type = "response")
y_pred_class_lr <- ifelse(y_pred_lr > 0.5, 1, 0)

# Convert predictions to factor
y_pred_class_lr <- factor(y_pred_class_lr, levels = c(0, 1))
Y_test <- factor(Y_test, levels = c(0, 1))

# Ensure levels match
levels(y_pred_class_lr) <- levels(Y_test)

# Get predicted probabilities on the test set
y_pred_proba_lr <- predict(model, newdata = X_test, type = "response")

# Print the first few predictions for verification
print(head(y_pred_class_lr))
print(head(y_pred_proba_lr))
```

```r
# Combine the training data
train_data <- data.frame(X_train, Class = Y_train)
```

```r
# Train the decision tree model
dt_model <- rpart(Class ~ ., data = train_data, method = "class", control = rpart.control(cp = 0.01))

# Predict class labels on the test set
y_pred_dt <- predict(dt_model, newdata = X_test, type = "class")

# Convert predictions to factor
y_pred_dt <- factor(y_pred_dt, levels = c(0, 1))

# Ensure levels match
levels(y_pred_dt) <- levels(Y_test)

# Predict probabilities on the test set
y_pred_proba_dt <- predict(dt_model, newdata = X_test, type = "prob")[, 2]

# Print the first few predictions for verification
print(head(y_pred_dt))
print(head(y_pred_proba_dt))
```

```r
# Combine the training data
train_data <- data.frame(X_train, Class = Y_train)
```

```r
# Convert 'Class' to a factor
train_data$Class <- as.factor(train_data$Class)

# Train the random forest model
set.seed(42)
rf_model <- randomForest(Class ~ ., data = train_data, importance = TRUE)

# Predict class labels on the test set
y_pred_rf <- predict(rf_model, newdata = X_test)

# Convert predictions to factor
y_pred_rf <- factor(y_pred_rf, levels = c(0, 1))

# Ensure levels match
y_pred_rf <- factor(y_pred_rf, levels = levels(Y_test))

# Predict probabilities on the test set
y_pred_proba_rf <- predict(rf_model, newdata = X_test, type = "prob")[, 2]

# Print the first few predictions for verification
print(head(y_pred_rf))
print(head(y_pred_proba_rf))
```

```r
# Calculate accuracies
accuracy_lr <- mean(y_pred_class_lr == Y_test)
accuracy_dt <- mean(y_pred_dt == Y_test)
accuracy_rf <- mean(y_pred_rf == Y_test)

# Print accuracies
cat("Logistic Regression Accuracy:", accuracy_lr, "\n")
cat("Decision Tree Accuracy:", accuracy_dt, "\n")
cat("Random Forest Accuracy:", accuracy_rf, "\n")
```

```r
# Generate confusion matrices for each model
cm_lr <- confusionMatrix(y_pred_class_lr, Y_test)
cm_dt <- confusionMatrix(y_pred_dt, Y_test)
cm_rf <- confusionMatrix(y_pred_rf, Y_test)

# Plot the confusion matrices
confusionMatrixPlot(cm_lr, "Confusion Matrix for Logistic Regression")
confusionMatrixPlot(cm_dt, "Confusion Matrix for Decision Tree")
confusionMatrixPlot(cm_rf, "Confusion Matrix for Random Forest")
```

```r
# Install the precrec package if not already installed
if (!requireNamespace("precrec", quietly = TRUE)) {
  install.packages("precrec")
}

library(precrec)

precisionRecall <- function(y_true, y_pred_proba, model_name) {
  # Ensure probabilities are numeric
  y_pred_proba <- as.numeric(y_pred_proba)

  # Create the precision-recall curve object
  pr_curve <- evalmod(scores = y_pred_proba, labels = y_true)

  # Plot the precision-recall curve
  plot(pr_curve, main = paste("Precision-Recall Curve for", model_name))
}
```

```r
# Generate precision-recall curve for Logistic Regression
pr_lr <- precisionRecall(Y_test, y_pred_proba_lr, "Logistic Regression")
print(pr_lr)

pr_dt <- precisionRecall(Y_test, y_pred_proba_dt, "Decision Tree")
print(pr_dt)

pr_rf <- precisionRecall(Y_test, y_pred_proba_rf, "Random Forest")
print(pr_rf)
```

```r
# Define function to compute precision from confusion matrix
compute_precision <- function(cm) {
  tp <- cm$table[2, 2]
  fp <- cm$table[1, 2]
  precision <- tp / (tp + fp)
  return(precision)
}

# Calculate precision for each model
precision_lr <- compute_precision(cm_lr)
precision_dt <- compute_precision(cm_dt)
precision_rf <- compute_precision(cm_rf)

# Print precisions
cat("Logistic Regression Precision:", precision_lr, "\n")
cat("Decision Tree Precision:", precision_dt, "\n")
cat("Random Forest Precision:", precision_rf, "\n")
```
